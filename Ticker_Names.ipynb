```python
import pandas as pd
import numpy as np
import string
```

```python
symbolsdf = pd.read_csv('NYSE.csv').append(pd.read_csv('NASDAQ.csv').append(pd.read_csv('AMEX.csv')))[['Symbol', 'Name']]
symbolsdf.head(10)
```

```python
symbolsdf.Name = symbolsdf.Name.str.translate(None, string.punctuation)
symbolsdf.Name = symbolsdf.Name.str.lower()
```

```python
droppeddf = pd.DataFrame(pd.read_csv('mutual funds 13-F.csv', index_col=0)['name of issuer'])
droppeddf.columns = ['Name']
droppeddf.Name = droppeddf.Name.str.translate(None, string.punctuation)
droppeddf.Name = droppeddf.Name.str.lower()
droppeddf
```

```python
orig_name = []
new_name = []
symbol = []
for orignial in droppeddf.Name:
    counter = 0
    for new in symbolsdf.Name:
        counter += 1
        if orignial == new:
            orig_name.append(orignial)
            new_name.append(new)
            symbol.append(symbolsdf.Symbol.iloc[counter])
            
        
```

```python
symbol_output = pd.DataFrame({'Original' : orig_name, 'Check' : new_name, 'Symbol' : symbol})
symbol_output
```

```python
symbol_output.Symbol.to_csv('Ticker_Symbols_list.csv')
```
